```python
!pip install pillow imageio matplotlib numpy -q
import numpy as np,matplotlib.pyplot as plt,json,io
from PIL import Image,ImageDraw,ImageFont
import imageio.v2 as imageio
print('Libraries loaded ✓')
```

```python
# NovaTech brand assets
LOGO_PATH='/content/drive/MyDrive/BrandSphere/data/novatech_logo.svg'
NOVATECH_PALETTE={'Onyx Black':'#1A1A1A','Electric Blue':'#5055D8','Deep Navy':'#0D1B3E','Warm Ivory':'#F9F9F7','Silver':'#C0C0C8'}
TAGLINE='Powering tomorrow, today.'
# Try to load previously generated taglines
try:
    with open('week4_output.json') as f: data=json.load(f); TAGLINE=data['taglines'][0]
except: pass
print(f'Using tagline: {TAGLINE}')
```

```python
def hex_to_rgb(h): return tuple(int(h[i:i+2],16) for i in (1,3,5))
def create_brand_frame(tagline,logo_scale,text_alpha,bg_color,text_color,width=800,height=600):
    img=Image.new('RGB',(width,height),hex_to_rgb(bg_color))
    draw=ImageDraw.Draw(img)
    # Logo area (falcon wing represented as geometric shape)
    cx,cy=width//2,height//2-60
    pts=[(cx,cy-60),(cx-140,cy+20),(cx-80,cy+20),(cx,cy+60),(cx+80,cy+20),(cx+140,cy+20)]
    scaled_pts=[(int(cx+(x-cx)*logo_scale),int(cy+(y-cy)*logo_scale)) for x,y in pts]
    draw.polygon(scaled_pts,fill=hex_to_rgb(text_color))
    # Tagline text
    try: font=ImageFont.truetype('/usr/share/fonts/truetype/dejavu/DejaVuSans.ttf',28)
    except: font=ImageFont.load_default()
    bbox=draw.textbbox((0,0),tagline,font=font); tw,th=bbox[2]-bbox[0],bbox[3]-bbox[1]
    tx,ty=(width-tw)//2,cy+90
    r,g,b=hex_to_rgb(text_color)
    draw.text((tx,ty),tagline,font=font,fill=(r,g,b,int(text_alpha*255)))
    # Wordmark
    try: wfont=ImageFont.truetype('/usr/share/fonts/truetype/dejavu/DejaVuSans.ttf',16)
    except: wfont=ImageFont.load_default()
    draw.text((width//2-45,cy+130),'NOVATECH',font=wfont,fill=hex_to_rgb(text_color))
    return np.array(img)
print('Frame generator ready ✓')
```

```python
# Build GIF animation (fade-in logo, then tagline)
frames=[]
n_frames=30
for i in range(n_frames):
    t=i/n_frames
    logo_scale=0.3+0.7*min(t*2,1.0)  # logo grows in first half
    text_alpha=max(0,min(1,(t-0.4)*2.5))  # text fades in second half
    frame=create_brand_frame(TAGLINE,logo_scale,text_alpha,'#F9F9F7','#1A1A1A')
    frames.append(frame)
# Hold last frame
frames.extend([frames[-1]]*10)
imageio.mimsave('novatech_brand_animation.gif',frames,fps=15,loop=0)
print('GIF saved: novatech_brand_animation.gif ✓')
# Preview first and last frames
fig,(a1,a2)=plt.subplots(1,2,figsize=(14,5))
a1.imshow(frames[0]); a1.set_title('Frame 1 — Start',fontweight='bold'); a1.axis('off')
a2.imshow(frames[-11]); a2.set_title('Frame 30 — Final',fontweight='bold'); a2.axis('off')
plt.tight_layout(); plt.savefig('week6_preview.png',dpi=150); plt.show()
```

```python
# Also create a colour-theme variation (dark)
dark_frames=[]
for i in range(n_frames):
    t=i/n_frames
    frame=create_brand_frame(TAGLINE,0.3+0.7*min(t*2,1.0),max(0,min(1,(t-.4)*2.5)),'#0D1B3E','#F9F9F7')
    dark_frames.append(frame)
dark_frames.extend([dark_frames[-1]]*10)
imageio.mimsave('novatech_brand_animation_dark.gif',dark_frames,fps=15,loop=0)
print('Dark variant GIF saved ✓')
```
